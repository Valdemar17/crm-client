# Presentation Slide Generation Strategy

We need to generate a "slide" (landscape PDF) for each credit request.

## Design Layout (Landscape 16:9 or 4:3)
- **Header**: SOFIMAS Logo (Top Left), "Comité de Crédito" (Top Right), Request ID/Date.
- **Title**: Client Name / Project Name (Large, centered or left).
- **Grid Layout**:
    - **Left Column**: Financial Terms (Monto, Tasa, Plazo, Comisión).
    - **Right Column**: Qualitative Data (Destino, Garantía, Observaciones).
- **Footer**: "Confidencial" status.
